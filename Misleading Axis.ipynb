When plotting a graph, you can make small differences seem large by choosing a scale that exaggerates the differences. For example, let's generate data for how much blobs like lattes, cappucinos, and americanos on average. We'll then set a misleading scale so make one coffee type seem much more popular than it is.

```python
import random as r
import matplotlib.pyplot as plt
r.seed(1)
```

```python
def generatePreferences():
    return([r.uniform(0, 1) for c in range(3)])

population = [generatePreferences() for c in range(1000)]
averages = [0 for c in range(3)]
for individual in population:
    for i in range(3):
        averages[i] += individual[i] / len(population)
        
labels = ["Latte", "Cappucino", "Americano"]
fig, axes = plt.subplots(1, 2, figsize = (10, 4))

axes[0].bar(labels, averages, color = ["#d4a373", "#6f4e37", "#8b4513"])
axes[0].set_ylabel("Mean Approval")
axes[0].set_ylim(min(averages) - 10 ** -3, max(averages) + 10 ** -3)
axes[0].set_title("Mean Coffee Approval (Misleading)")

axes[1].bar(labels, averages, color=["#d4a373", "#6f4e37", "#8b4513"])
axes[1].set_ylabel("Mean Approval")
axes[1].set_ylim(0, 1)
axes[1].set_title("Mean Coffee Approval (Fair)")
plt.tight_layout()
plt.show()
```
