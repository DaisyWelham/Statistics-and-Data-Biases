Suppose we have 100,000 blobs that we allocate to different sized towns. There are 200 towns of size 100, 50 of size 250, 50 of 500, 30 of 750, and 20 of 1,000. Each blob has a 40% chance of liking avocados, regardless of which town they live in. 

Due to sheer chance, small towns can have unusually high or unusually low rates of liking avocado. This can make it tempting to seek a causal relationship between small town size and high/low rates of liking avocado, but this is fallacious because there is no causal relationship.

```python
import random as r
import matplotlib.pyplot as plt
r.seed(1)
```

Data

```python
def generateTown(townSize):
    output = []
    for i in range(townSize):
        output.append(1 if r.uniform(0, 1) < 0.4 else 0)
    output = [output, sum(output) / len(output), len(output)]
    return(output)

allTowns = []
townSizes = [100, 250, 500, 750, 1000]
townNumbers = [200, 50, 50, 30, 20]
for i in range(5):
    for j in range(townNumbers[i]):
        allTowns.append(generateTown(townSizes[i]))

for town in allTowns:
    plt.scatter(town[2], town[1])
plt.xlim(0, 1025)
plt.ylim(0.2, 0.6)
plt.grid(True)
plt.title("Town size versus avocado approval rate")
plt.ylabel("Avocado approval ratio")
plt.xlabel("Town size (Blobs)")
plt.show()
```
