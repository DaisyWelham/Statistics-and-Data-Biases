Volunteer bias occurs when a sample is self-selecting in some way. People are more likely to volunteer for things they like and/or are good at, so self-selecting samples tend to view the things they volunteer for more favourably and be better at those things than a more representative sample.

In our example, there is a school of 1000 blobs and we need volunteers for sports day. Each blob has a score for athleticism and enthusiasm, and blobs are more likely to volunteer for sports day if they are both athletic and enthusiastic. We'll also track the blobs' fondness for lattes, to show that the average only changes if blobs are more likely to volunteer based on it. We'll see that our average athleticism from among the sports day volunteers is much higher than from the population of blobs as a whole.

```python
import random as r
import matplotlib.pyplot as plt
r.seed(1)
```

```python
def generateIndividual():
    athleticism = r.uniform(0, 1)
    enthusiasm = r.uniform(0, 1)
    lattes = r.uniform(0, 1)
    volunteers = 1 if r.uniform(0, 1) < (athleticism * enthusiasm) else 0
    return([[athleticism, enthusiasm, lattes], volunteers])

population = [generateIndividual() for c in range(1000)]

colors = ["Red", "Green"]
for individual in population:
    plt.scatter(individual[0][0], individual[0][1], color = colors[individual[1]], marker = "x")
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.title("Blob Athleticism & Enthusiasm versus whether they volunteered (green/red)")
plt.xlabel("Athleticism")
plt.ylabel("Enthusiasm")
plt.show()

volunteers = [individual for individual in population if individual[1] == 1]

averageAthleticism = 0
volunteerAverageAthleticism = 0
averageLatte = 0
volunteerAverageLatte = 0
for individual in population:
    averageAthleticism += individual[0][0] / len(population)
    averageLatte += individual[0][2] / len(population)
    if individual[1] == 1:
        volunteerAverageAthleticism += individual[0][0] / len(volunteers)
        volunteerAverageLatte += individual[0][2] / len(volunteers)

print(f"Whole group averages: Athleticism: {averageAthleticism}, Latte Preference: {averageLatte}")
print(f"Volunteer averages: Athleticism: {volunteerAverageAthleticism}, Latte Preference: {volunteerAverageLatte}")
```

We see that athleticism increases significantly in the volunteer population compared to the whole population since the volunteers are deciding partly on the basis of their athleticism. The latte preference doesn't change much because blobs who like lattes are not more likely to volunteer.
